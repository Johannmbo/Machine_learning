```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.diagnostic import lilliefors
import statsmodels.api as sm
```

```python
# Let's upload the dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Let's select the interest value
sepal_length = iris['sepal_length']
```

```python
# Skewness and kurtosis values
skewness = stats.skew(sepal_length)
kurtosis = stats.kurtosis(sepal_length, fisher=False) #True kurtosis
excess_kurtosis = stats.kurtosis(sepal_length)
print('Skewness:', skewness)
print('Kurtosis:', kurtosis)
print('Excess kurtosis:', excess_kurtosis)
```

## Graphics analysis

```python
#Histogram
plt.figure()
sns.histplot(sepal_length, kde=True)
plt.title('Histogram for sepal_length')
plt.xlabel('sepal_length')
plt.ylabel('Frequency')
plt.show()
```

```python
#Boxplot
plt.figure()
sns.boxplot(x = sepal_length)
plt.title('Boxplot for sepal_length')
plt.xlabel('sepal_length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Q-Q plot
plt.figure()
stats.probplot(sepal_length, dist='norm', plot=plt)
plt.title('Q-Q plot for sepal_length')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Sample quantiles')
```

```python
# Niveau de signification (alpha)
alpha = 0.05

# Effectuer les tests de normalite
shapiro_test = stats.shapiro(sepal_length)
ks_test = stats.kstest(sepal_length, 'norm')
ad_test = stats.anderson(sepal_length, dist='norm')
dagostino_test = stats.normaltest(sepal_length)
lilliefors_test = lilliefors(sepal_length)

# Creer un tableau pandas avec les resultats des tests
test_results = pd.DataFrame({
    'Test name': ['Shapiro-Wilk', 'Kolmogorov-Smirnov', 'Anderson-Darling', 'Dagostino-Pearson', 'Lilliefors'],
    'test statistic': [shapiro_test[0], ks_test.statistic, ad_test.statistic, dagostino_test.statistic, lilliefors_test[0]],
    'p-value': [shapiro_test[1], ks_test.pvalue, None, dagostino_test.pvalue, lilliefors_test[1]],
    'Normalite': ['yes' if shapiro_test[1] > alpha else 'no', 'yes' if ks_test.pvalue > alpha else 'no', None, 'yes' if dagostino_test.pvalue > alpha else 'no', 'yes']
})
print (test_results)
```
