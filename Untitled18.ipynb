```python
from sklearn.cluster import KMeans
k = 2
kmeans = KMeans(n_clusters=k)
labels = kmeans.fit_predict(data[['x','y']])

print("Nombre de clusters =", len(set(labels)))
```

```python
from sklearn.cluster import KMeans

inertias = []
K = range(1, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data[['x','y']])
    inertias.append(kmeans.inertia_)

plt.plot(K, inertias, marker='o')
plt.xlabel("k (nombre de clusters)")
plt.ylabel("Inertia")
plt.title("Mthode du coude")
plt.show()
```

```python
train_iris = pd.read_csv("/content/train_iris.csv")
```

```python
test_iris = pd.read_csv("/content/test_iris.csv")
```

```python
test_iris.drop('Unnamed: 0', axis = 1)
```

```python
train_iris = train_iris.drop('Unnamed: 0', axis = 1)
```

```python
train_iris
```

```python
train_iris.isnull().sum()
```

```python
X_train = train_iris[['sepal length', 'sepal width', 'petal length', 'petal width']]
y_train = train_iris['species']
```

```python
X_test = test_iris[['sepal length', 'sepal width', 'petal length', 'petal width']]
y_test = test_iris['species']
```

```python
import numpy as np

rs = np.random.seed(0)
```

```python
my_awesome_tree = tree.DecisionTreeClassifier(criterion = 'entropy', random_state = rs)
```

```python
max_depth_values = range(1, 100)
```

```python
my_awesome_tree.fit(X_train, y_train)
```

```python
my_awesome_tree.score(X_train, y_train)
```

```python
cross_val_score(my_awesome_tree, X_train, y_train, cv=5)
```

```python
y_pred = my_awesome_tree.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
scores_data = pd.DataFrame()
```

```python
data = []
```

```python
for max_depth in max_depth_values:
    my_awesome_tree = tree.DecisionTreeClassifier(criterion = 'entropy', max_depth = max_depth, random_state = rs)
    my_awesome_tree.fit(X_train, y_train)
    train_score = my_awesome_tree.score(X_train, y_train)
    y_pred = my_awesome_tree.predict(X_test)
    test_score = accuracy_score(y_test, y_pred)
    mean_cross_val_score = cross_val_score(my_awesome_tree, X_train, y_train, cv=5).mean()

    temp_data = (max_depth, train_score, test_score, mean_cross_val_score)
    data.append(temp_data)
    scores_data = pd.DataFrame(data, columns=['max_depth', 'train_score', 'test_score', 'mean_cross_val_score'])
```

```python
scores_data.head()
```

```python
scores_data_long = scores_data.melt(id_vars=['max_depth'], value_vars=['train_score', 'test_score', 'mean_cross_val_score'],
        var_name='set_type', value_name='score')
scores_data_long.head()
```

```python
sns.lineplot(data=scores_data_long, x="max_depth", y="score", hue="set_type")
```
